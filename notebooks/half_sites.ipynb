```python
import pandas as pd
import genome_utils
import genomepy

genomic_reference_file = '../data/references/hg38.fa'

# Load cryptic sites
sites = pd.read_csv('./data/gt_cryptic.csv')
sites = genome_utils.fetch_genomic_sites(sites, genomic_reference_file)
```

```python
# Split into left and right
complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
reverse_complement = lambda x: "".join(complement.get(base, base) for base in reversed(x))

left = sites['sequence'].str.slice(0,22)
left_rc = left.apply(reverse_complement)
right = sites['sequence'].str.slice(24,)
right_rc = right.apply(reverse_complement)

sites['left_sym'] = left + 'GT' + left_rc
sites['right_sym'] = right_rc + 'GT' + right
```

```python
# attB sequence
attB = 'GGCCGGCTTGTCGACGACGGCGGTCTCCGTCGTCAGGATCATCCGG'
zr_left = attB[0:8]
zr_right = attB[-8:]

zr_left_rc = reverse_complement(zr_left)
zr_right_rc = reverse_complement(zr_right)

sites['left_sym_left_zr'] = zr_left + sites['left_sym'].str.slice(8,-8) + zr_left_rc
sites['left_sym_right_zr'] = zr_right + sites['left_sym'].str.slice(8,-8) + zr_right_rc
sites['right_sym_left_zr'] = zr_left + sites['right_sym'].str.slice(8,-8) + zr_left_rc
sites['right_sym_right_zr'] = zr_right + sites['right_sym'].str.slice(8,-8) + zr_right_rc
```

```python
sites['strand'] = sites['positive_strand'].apply(lambda x: '+' if x else '-')
sites['short_name'] = sites['reference_name'] + ':' + sites['position'].astype(str) + ':' + sites['strand']
sites
```
